```python
import numpy as np
```

```python
xs = [10, 20, 30, 40, 100, 200]
ys = [20, 30, 40, 50, 200, 300]

xys = [[10, 20], [20, 30], [30, 40], [40, 50], [100, 200], [200, 300]]
xys = np.array(xys)
```

```python
print(xys)
```

```python
plt.scatter(xys[ :,0], xys[ :,1])
```

```python
mean = np.mean(xys, 0)
print(mean)
```

```python
import matplotlib.patches as patches

plt.scatter(xys[ :,0], xys[ :,1])
plt.scatter(xys, 0)

mean = np.mean(xys, 0)
std_dev = np.std(xys[ :,0], xys[ :,1])

ellipse = patches.Ellipse([mean[0], [mean[1]], std_dev[0]*2, std_dev[0]*2, alpha=0.25)

fig,graph = plt.subplot()

graph.scatter(xys[ :,0], xys[ :,1])
graph.scatter(xys, 0)
graph.add_patche(ellipse)
```
